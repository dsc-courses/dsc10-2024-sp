```python
import os
import re
import csv
import requests
from PIL import Image
from io import BytesIO

def download_and_save_images(csv_content, directory):
    # Create directory if it doesn't exist
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    # Parse the CSV content line by line
    reader = csv.reader(csv_content.splitlines())
    #next(reader)  # Skip the header row
    
    for row in reader:
        print(row)
        timestamp, full_name, pronouns, link = row[0].split("\t")
        first_name, last_name = full_name.split()
        
        # Format the file name
        file_name = f"{last_name.lower()}-{first_name.lower()}.jpg"
        file_path = os.path.join(directory, file_name)
        
        match = re.search(r'/d/([^/]+)/', link)
        if match:
            file_id = match.group(1)
            download_link = f"https://drive.google.com/uc?export=download&id={file_id}"
        else:
            print(f"Failed to extract file ID from link for {full_name}. Skipping.")
            continue
        
        try:
            # Download the image using the modified download link
            response = requests.get(download_link)
            response.raise_for_status()
            
            # Open the image and convert it to JPEG if necessary
            img = Image.open(BytesIO(response.content))
            if img.format != 'JPEG':
                img = img.convert('RGB')
            
            # Save the image to the specified directory
            img.save(file_path, 'JPEG')
            print(f"Image for {full_name} has been saved successfully to {file_path}")
        
        except Exception as e:
            print(f"Failed to download and save image for {full_name}. Error: {e}")

# Copy-paste the first four columns from the staff onboarding form responses
csv_content = """3/22/2024 22:03:53	Jason Huynh	he/him	https://drive.google.com/file/d/106jM7_GJ_JYm_G2-i5Y7iiV56CfkkAcB/view?usp=sharing
3/23/2024 6:39:08	Cici Xu	she/her	https://drive.google.com/file/d/1vlwXr1SzpOnw6DJB7E2nlScMiEEbO0Qp/view?usp=drive_link
3/23/2024 12:28:14	Kate Feng	she/her	https://drive.google.com/file/d/1mxx36mI_uuCd1momzQEdtBwRXveeCvFS/view?usp=sharing
3/23/2024 12:43:14	Athulith Paraselli	he/him	https://drive.google.com/file/d/1Qv6mOHmhxc-OiCUIJgz-LAdCm6IxqtC_/view?usp=drivesdk
3/23/2024 16:30:43	Ashley Ho	she/her	https://drive.google.com/file/d/1axBMlAyfSICSAP0nZZtjnPi6-aEkZ0in/view?usp=sharing
3/23/2024 20:49:23	Claire Wang	she/her	https://drive.google.com/file/d/1K5mg4mE6ytpYYhK4MDDTBLcfjK7Y1A0Y/view?usp=sharing
3/24/2024 2:05:28	Yutian Shi	she/her	https://drive.google.com/file/d/1_6Wf7I7fNSvT3F7FqZGLze1g1ZkOtSNC/view?usp=share_link
3/24/2024 2:40:09	Oren Ciolli	he/him	https://drive.google.com/file/d/10KWQ5rUMgfQT_L-Af4452BST5o97dflb/view?usp=drive_link
3/24/2024 22:31:09	Pallavi Prabhu	she/her	https://drive.google.com/file/d/1ohZshbfRa9V2A-tACoXrtSYaCTc-oWX6/view?usp=share_link
3/25/2024 11:27:50	Sofia Tkachenko	she/her	https://drive.google.com/file/d/12_B5BDtQCgSzToeFfoygW_WedO_cx5sn/view?usp=drivesdk
3/22/2024 20:19:58	Aaron Rasin	he/him	https://drive.google.com/file/d/1pZSLFL9gGeMK-ND-z5FhDcMI81me3VPw/view?usp=sharing
3/22/2024 21:15:29	Minchan Kim	he/him	https://drive.google.com/file/d/1euPRK57DMQWbURWkitbomL7Rn8My8Iqn/view?usp=sharing
3/22/2024 22:11:01	Sophie Wang	she/her	https://docs.google.com/document/d/1JrYLMl5SnxqZ_DYmmnUTYlUXnWLLNy243v37cc8pMEU/edit?usp=sharing
3/23/2024 15:12:37	Pranav Rajaram	he/him	https://drive.google.com/file/d/1_A2ja6xNUg_9-Dz2ehCl7njsZ4aQ82Rg/view?usp=sharing
3/23/2024 16:02:58	Daniel Budidharma	he/him	https://drive.google.com/file/d/1xwsVBFTvo6KaiobE2BEn9KSWhhn9SXnS/view?usp=sharing
3/23/2024 22:31:10	Norah Kerendian	she/her	https://drive.google.com/file/d/12qunEaotQ4Czbk3WPJNxJKWkud84HwnQ/view?usp=sharing
3/23/2024 22:47:22	Calvin Nguyen	he/him	https://drive.google.com/file/d/1zM9dpb4u2CBMSN8FUuc-T9DfIMT2Zmsw/view?usp=sharing
3/23/2024 23:16:10	Raine Hoang	they/them	https://drive.google.com/file/d/1-Lxj8yY1REXiJym1caTnncbCFYi4O5RR/view?usp=sharing
3/24/2024 0:16:50	Michelle Hong	she/her	https://drive.google.com/file/d/1lQYR70tVWKExJzX9J79EXnXd9f7f_xxJ/view?usp=sharing
3/25/2024 1:51:30	Ciro Zhang	he/him	https://drive.google.com/file/d/1sw0lE791jZmE3fRRtuHlT9a8ibpZ2pWi/view?usp=sharing
3/25/2024 12:26:08	Vivian Lin	she/her	https://drive.google.com/file/d/14uJ06sY3W2xoOva256fEfbrECGweFtaA/view?usp=drivesdk"""

# The directory where the images should be saved
directory = "../assets/staff-images/sp24"

# Call the function with the CSV content and the directory
download_and_save_images(csv_content, directory)

# This will extract some, others will need to be done manually.
```
